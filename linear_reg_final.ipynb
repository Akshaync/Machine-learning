```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
x_train= train_data[['GrLivArea','FullBath','HalfBath']]
y_train=train_data['SalePrice']

x_test=dxtest_data[['GrLivArea','FullBath','HalfBath']]
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

x_train_split, x_val_split, y_train_split, y_val_split = train_test_split(
    x_train_scaled, y_train, test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train_split,y_train_split)
y_val_pred=model.predict(x_val_split)
```

```python
mse = mean_squared_error(y_val_split, y_val_pred)
print(f'Mean Squared Error on Validation Set: {mse}')
```

```python
y_test_pred = model.predict(x_test_scaled)

```

```python
plt.scatter(y_val_split, y_val_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices on Validation Set')
plt.show()
```

```python
plt.scatter(y_test, y_test_pred)
plt.plot(y_test,y_test_pred, color="black",linewidth=3)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices on Test Set')
plt.savefig('actual_vs_predicted_test_set.png')
plt.show()
```

```python
output_df = pd.DataFrame({'Actual Prices': y_test, 'Predicted Prices': y_test_pred})

# Save the DataFrame to a CSV file
output_df.to_csv(r"C:\Users\aksha\OneDrive\Desktop\prefinal.csv", index=False)
```
